```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('review.xlsx');
plt.figure(figsize=(14, 8))
plt.plot(df.iloc[:, 0], df.iloc[:, 1], label=df.columns[1], marker='.')
plt.plot(df.iloc[:, 0], df.iloc[:, 2], label=df.columns[2], marker='.')
plt.xlabel('Число потоков')
plt.ylabel('Время выполнения (секунды)')
plt.legend()
plt.grid()
plt.show()
```

```python
df = pd.read_excel('review.xlsx');
plt.figure(figsize=(14, 8))
plt.plot(df.iloc[:, 0], df.iloc[:, 3], label=df.columns[3], marker='.')
plt.xlabel('Число потоков')
plt.ylabel('Время выполнения (секунды)')
plt.legend()
plt.grid()
plt.show()
```
